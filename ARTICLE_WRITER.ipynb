Imports

```python
import warnings
import os
from crewai import Agent, Task, Crew
from IPython.display import Markdown
from google.colab import userdata
from langchain_openai import ChatOpenAI
warnings.filterwarnings('ignore')

```

```python
openai_api_key = userdata.get("OPENAI_API_KEY")
if not openai_api_key:
    raise ValueError("Openai API Key not found. Make sure it's set in Colab secrets under DEEPSEEK_API_KEY.")

# Set DeepSeek API Key for model interaction
os.environ["OPENAI_API_KEY"] = openai_api_key
os.environ["OPENAI_API_BASE"] = 'gpt-3.5-turbo'


```

Creating Agents

AGENT PLANNER
The planner is responsible for structuring the content, ensuring the article is logically organized and addresses key questions that a non-medical audience might have.

```python
planner = Agent(
    role="Content Planner",
    goal="Create an easy-to-understand content plan on Alzheimer's disease for non-healthcare readers.",
    backstory="You are a researcher who specializes in simplifying medical topics for the general public."
              " Your goal is to create a structured, engaging outline that breaks down complex concepts in a way that"
              " anyone can understand.",
    allow_delegation=False,
    verbose=True,
```

AGENT WRITER
The writer transforms the planner's outline into a compelling article,using accessible language and relatable storytelling to keep the reader engaged.

```python
writer = Agent(
    role="Content Writer",
    goal="Write a clear, engaging, and simple article about Alzheimer's disease for non-experts.",
    backstory="You are a writer with a talent for explaining medical topics in an engaging, easy-to-read format."
              " Your goal is to ensure the article is approachable and free from unnecessary medical jargon."
              " Use analogies, storytelling, and real-life examples to keep the reader engaged.",
    allow_delegation=False,
    verbose=True,

)
```

AGENT EDITOR
The editor ensures the final article is polished, easy to understand,and free from medical jargon. This role also focuses on formatting and structuring the content for optimal readability.

```python
editor = Agent(
    role="Editor",
    goal="Ensure readability and accessibility of the article for a non-medical audience.",
    backstory="You are an editor who specializes in making complex topics simple and engaging."
              " Your task is to proofread the article, simplify difficult terms, and ensure a smooth reading experience."
              " The article should be structured well and have a friendly, conversational tone.",
    allow_delegation=False,
    verbose=True,

)
```

Define Tasks

PLANNERS TASK
The planner's task involves breaking down Alzheimer's disease into digestible sections, identifying common questions, and ensuring the content is structured logically.

```python
plan_task = Task(
    description=(
        "1. Research Alzheimer's disease and break it down into easy-to-understand sections.\n"
        "2. Identify key questions that non-experts might have about the disease.\n"
        "3. Create an engaging content outline, including real-world examples or personal stories.\n"
        "4. List key points in a logical order, ensuring clarity and flow."
    ),
    expected_output="A structured content plan that outlines the main topics, key questions, and engaging examples.",
    agent=planner,
)
```

WRITERS TASK
The writer expands on the planner's outline, ensuring the article remains engaging and understandable while integrating storytelling and relatable examples.

```python
write_task = Task(
    description=(
        "1. Use the content plan to write an engaging article on Alzheimer's disease.\n"
        "2. Explain medical concepts using everyday language.\n"
        "3. Use storytelling or relatable examples to make it easier to understand.\n"
        "4. Ensure a friendly and encouraging tone, avoiding fear-inducing language."
    ),
    expected_output="A well-written, easy-to-read article that educates and engages the audience.",
    agent=writer,
)
```

EDITORS TASK
The editor refines the final article by checking for readability,simplifying complex terms, ensuring smooth transitions, and structuring the content for better engagement.

```python
edit_task = Task(
    description=(
        "1. Proofread the article for readability and simplicity.\n"
        "2. Replace or explain any complex medical terms.\n"
        "3. Ensure smooth transitions and engaging storytelling.\n"
        "4. Format the article with short paragraphs and clear headings."
    ),
    expected_output="A polished, easy-to-understand article ready for publication.",
    agent=editor,
)
```

CREATE CREW  
The crew consists of the three agents working together in sequence:
The planner structures the content, the writer crafts the article, and the editor refines it for clarity and readability.

```python
crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan_task, write_task, edit_task],
    verbose=2
)
```

Run the Crew  
The execution starts with a predefined topic, but it can be changed dynamically to generate articles on different aspects of Alzheimer's disease.

```python
result = crew.kickoff(inputs={"topic": "Understanding Early Signs of Alzheimer's Disease"})
Markdown(result)
```
